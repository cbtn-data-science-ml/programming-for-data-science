# Import and Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Raw Dataset

```python
url = 'https://raw.githubusercontent.com/jonathan-data-analysis/colab_test_data/main/athletes.csv'
df = pd.read_csv(url) # convert to dataframe
```

## Class Object

```python
class Summary:
    def __init__(self, data):
        self.df = pd.read_csv(data)
    def print_head(self, rows=None): # rows=None is for optional args
        print(self.df.head(rows))
    def get_columns(self):
        print(list(self.df.columns))
    def get_dim(self):
        print('Rows:', len(self.df))
        print('Columns:', len(list(self.df.columns)))
    def get_stats(self):
        print(self.df.describe())
    def get_mean(self, column):
        print(f"Mean {column}:", self.df[column].mean())
    def get_standard_dev(self, column):
        print(f"STD {column}:", self.df[column].std())
    def get_hist(self, column): # data viz!!
        sns.set()
        self.df[column].hist(bins=10)
        plt.title(f"{column} Histogram")
        plt.show()
```

## Call Method Using a Variable Name `data`

```python
# assign class with raw_data to varible name data
data = Summary(url)
```
